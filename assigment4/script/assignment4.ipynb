```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
path = "../data/iris.data.head.csv"
data =  pd.read_csv(path)
iris = pd.DataFrame(data)
iris
```

```python
s_length = iris["sepal_length"].values
s_width = iris["sepal_width"].values
p_length = iris["petal_length"].values
p_width = iris["petal_width"].values
```

```python
plt.hist(s_length)
plt.title("Sepal_length")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid()
plt.show
```

```python
plt.hist(s_width)
plt.title("Sepal_width")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid()
plt.show
```

```python
plt.hist(p_length)
plt.title("Petal_length")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid()
plt.show
```

```python
plt.hist(p_width)
plt.title("Petal_width")
plt.xlabel("value")
plt.ylabel("frequency")
plt.grid()
plt.show
```

```python
#Standardized Z (respects outliers)
s_sl = (s_length -np.mean(s_length))/np.std(s_length)
s_sw = (s_width -np.mean(s_width))/np.std(s_width)
s_pl = (p_length -np.mean(p_length))/np.std(p_length)
s_pw = (p_width -np.mean(p_width))/np.std(p_width)
plt.hist(s_pl)
```

```python
#Normalized (moves all values more to the mean)
n_sl = (s_length - min(s_length))/max(s_length)
n_sw = (s_width - min(s_width))/max(s_width)
n_pl = (p_length - min(p_length))/max(p_length)
n_pw = (p_width - min(p_width))/max(p_width)
```

```python
#Logarithm transformation (to undo asymmetry towards the right around mean)
pw_log = np.log(p_width)
pw_log_s = (pw_log - min(pw_log))/max(pw_log)
pw_log_n = (pw_log - np.mean(pw_log))/ np.std(pw_log)
plt.hist(pw_log_s, alpha = 0.6)
#normalization shifts it to minus 2
plt.hist(pw_log_n, alpha = 0.6)
# log norm doesn't seem to make much sense since the values are not that far apart
```

```python
#If conditions to separate data into 3 classes
iris["class"][::50]
is_setosa = iris["class"] == "Iris-setosa"
is_virg = iris["class"] == "Iris-virginica"
is_versi = iris["class"] == "Iris-versicolor"
```

```python
#Sepal length vs Sepal Width
plt.scatter(s_sl[is_setosa], n_sw[is_setosa])
plt.scatter(s_sl[is_virg], n_sw[is_virg])
plt.scatter(s_sl[is_versi], n_sw[is_versi])
plt.title("Sepal length vs Sepal Width")
plt.xlabel("sepal_length")
plt.ylabel("sepal_width")
plt.legend(["Setosa","Virginica","Versicolor"])
```

```python
#Sepal length vs Petal Length
plt.scatter(s_sl[is_setosa], n_pl[is_setosa])
plt.scatter(s_sl[is_virg], n_pl[is_virg])
plt.scatter(s_sl[is_versi], n_pl[is_versi])
plt.title("Sepal length vs Petal Length")
plt.xlabel("sepal_length")
plt.ylabel("petal_length")
plt.legend(["Setosa","Virginica","Versicolor"])
```

```python
#Sepal length vs Petal Width
plt.scatter(s_sl[is_setosa], n_pw[is_setosa])
plt.scatter(s_sl[is_virg], n_pw[is_virg])
plt.scatter(s_sl[is_versi], n_pw[is_versi])
plt.title("Sepal length vs Petal Width")
plt.xlabel("sepal_length")
plt.ylabel("petal_width")
plt.legend(["Setosa","Virginica","Versicolor"])
```

```python
#Sepal Width vs Petal length
plt.scatter(n_sw[is_setosa], n_pl[is_setosa])
plt.scatter(n_sw[is_virg], n_pl[is_virg])
plt.scatter(n_sw[is_versi], n_pl[is_versi])
plt.title("Sepal Width vs Petal Length")
plt.xlabel("sepal_width")
plt.ylabel("petal_length")
plt.legend(["Setosa","Virginica","Versicolor"])
```

```python
#Sepal Width vs Petal Width
plt.scatter(n_sw[is_setosa], n_pw[is_setosa])
plt.scatter(n_sw[is_virg], n_pw[is_virg])
plt.scatter(n_sw[is_versi], n_pw[is_versi])
plt.title("Sepal Width vs Petal Width")
plt.xlabel("sepal_width")
plt.ylabel("petal_width")
plt.legend(["Setosa","Virginica","Versicolor"])
```
