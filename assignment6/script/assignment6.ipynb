```python
# dataset: face.csv
# perform and visualize clustering
# initial inspection (scatterplots)
# choice of number of clusters (elbow method)
# visualize clusters only with the second and third variable
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
face_df = pd.read_csv("../data/face.csv", index_col = 0)
plt.scatter(face_df.iloc[:,0], face_df.iloc[:,1])
```

```python
# elbow method how many clusters to optimize the number of clusters
dist =[]
for i in range(1,20):
    km = cluster.KMeans(n_clusters=i, init="k-means++")
    km.fit(face_df)
    dist.append(km.inertia_)

plt.plot(range(1,20),dist,"o-")
plt.show()
```

```python
km = cluster.KMeans(n_clusters=6, init="k-means++")
face_df_c = km.fit_predict(face_df)
```

```python
print("Labels : ",km.labels_)
print("Centers : ",km.cluster_centers_)
print("Inertia : ",km.inertia_)
print("Clusters : ",face_df_c)
```

```python
plt.scatter(face_df.iloc[face_df_c ==0, 0],face_df.iloc[face_df_c ==0, 1], c = "purple", marker = "x", label = "cluster 1")
plt.scatter(face_df.iloc[face_df_c ==1, 0],face_df.iloc[face_df_c ==1, 1], c = "darkblue", marker = "o", label = "cluster 2")
plt.scatter(face_df.iloc[face_df_c ==2, 0],face_df.iloc[face_df_c ==2, 1], c = "pink", marker = "v", label = "cluster 3")
plt.scatter(face_df.iloc[face_df_c ==3, 0],face_df.iloc[face_df_c ==3, 1], c = "orange", marker = "s", label = "cluster 4")
plt.scatter(face_df.iloc[face_df_c ==4, 0],face_df.iloc[face_df_c ==4, 1], c = "red", marker = "s", label = "cluster 5")

# Tested up to 9 clusters but it starts looking rather confusing after 6 clusters.
plt.scatter(face_df.iloc[face_df_c ==5, 0],face_df.iloc[face_df_c ==5, 1], c = "green", marker = "o", label = "cluster 6")
#plt.scatter(face_df.iloc[face_df_c ==6, 0],face_df.iloc[face_df_c ==6, 1], c = "aqua", marker = "o", label = "cluster 7")
#plt.scatter(face_df.iloc[face_df_c ==7, 0],face_df.iloc[face_df_c ==7, 1], c = "yellow", marker = "o", label = "cluster 8")
#plt.scatter(face_df.iloc[face_df_c ==8, 0],face_df.iloc[face_df_c ==8, 1], c = "k", marker = "o", label = "cluster 9")

plt.legend()
plt.title("clusters of face.csv")
plt.legend(loc= "center left", bbox_to_anchor = [1,.5])
plt.grid()
plt.show()

```
